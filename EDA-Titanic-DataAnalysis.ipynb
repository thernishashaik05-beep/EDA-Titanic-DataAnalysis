```python
# Task 2: Exploratory Data Analysis (EDA) - Titanic Dataset

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\thern\OneDrive\Desktop\ElevateLabs\Titanic_Cleaned.xls"
)

print(df.head())
```

```python
 #Basic Exploration

print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Information:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Summary Statistics

print("\nSummary Statistics:")
print(df.describe())

print("\nMedian Values:")
print(df.median(numeric_only=True))
```

```python
 #Histograms for Numeric Features

numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
#Boxplots for Numeric Features

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
 #Survival Distribution

plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
#Survival by Gender

plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
 #Survival by Passenger Class

plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
#Pairplot

pairplot_data = df[['Survived', 'Pclass', 'Age', 'Fare']].dropna()

sns.pairplot(pairplot_data, hue='Survived')
plt.show()
```

```python
#Correlation Matrix
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
#Basic Insights

print("\n----- INSIGHTS -----")

print("1. Female passengers had a higher survival rate than males.")
print("2. First-class passengers survived more frequently.")
print("3. Fare contains several outliers.")
print("4. Age has some missing values.")
print("5. Passenger class and survival are related.")
print("6. Most passengers were between 20 and 40 years old.")
```
